### Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Importing csv:

```python
ed = pd.read_csv('../datasets/education_cleaned.csv', index_col = 0, sep = ';')
```

```python
hp = pd.read_csv('../datasets/happiness_cleaned.csv', index_col = 0)
```

### I want to see if some of the top 30 happiest countries are also the ones with a higher level of education:

```python
# top30 most educated 
ed_top30 = ed.groupby("Country").sum().sort_values("EducationIndex", ascending=False)[0:30]
```

```python
ed_top30
```

```python
# top10 happiest:
happiest = hp.groupby("Country").sum().sort_values("Happiness_Score", ascending=False)[0:30]
```

```python
happiest
```

```python
# Checking the countries that appear in both:

for x in ed_top30.index:
    if x in happiest.index:
        print(x)
```

### As we can see 17 out of the 30 countries appear in both so it can mean there is some relation between the education index and how happy a country is. let's check if the botton 30 are also countries with a lower level of education:

```python
unhappiest = hp.groupby("Country").sum().sort_values("Happiness_Score")[1:31]
```

```python
unhappiest
```

```python
ed_botton30 = ed.groupby("Country").sum().sort_values("EducationIndex")[0:30]
```

```python
ed_botton30
```

```python
for x in ed_botton30.index:
    if x in unhappiest.index:
        print(x)
```

### As we can see 50% of the botton 30 countries with a lower index of education are also the unhappiest countries. Let's see the correlation between this two variables:

```python
combined = pd.merge(ed, hp, on = 'Country', how = 'inner')
```

```python
combined = combined[['Country', 'EducationIndex', 'Happiness_Score']]
```

```python
combined["EducationIndex"].corr(combined["Happiness_Score"])
```

```python
combined["Happiness_Score"].isnull().sum()
```

### Correlation between Education Index and Happiness_Score is 0.77, so we can appreciate a high correlation between variables.

```python
import statsmodels.api as sm
```

```python
X = sm.add_constant(combined['EducationIndex']) # variable independiente
model = sm.OLS(combined['Happiness_Score'],X) # variable dependiente
results = model.fit()
print(results.summary())
```
