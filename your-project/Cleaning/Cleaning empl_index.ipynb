```python
empl = pd.read_csv('../Cleaning/UNEMPLOYMENT RATE 2017.csv',sep=';', index_col = 0)
```

```python
empl = empl.reset_index()
```

```python
empl
```

```python
# I'm going to create an index of employment where:
# If the unemplyment rate is higher than 22%, the index will be 0.
# If the unemployment is higher than 15& the index will be 0.25.
# If the unemployment is higher or equal to 10%, the index will be 0.5.
# If the unemployment is higher than 6, the index will be 0.75.
# If the unemployment is higher lower than 6, the index will be a 1.
```

```python
def assig_values(x):
    if x['Unemployment_rate'] < 6:
        return 1
    elif x['Unemployment_rate'] <= 10:
        return 0.75
    elif x['Unemployment_rate'] < 15:
        return 0.5
    elif x['Unemployment_rate'] < 22:
        return 0.25
    else:
        return 0
```

```python
empl['Index_employment'] = empl.apply(assig_values, axis = 1)
```

```python
empl = empl[['Country', 'Index_employment']]
```
